#### Introduction: <br>
With the rise of solar panels, a debate has risen on how renewable these technologies are. Although helping reduce emissions from electricity, critics claim that solar panels need to be manufactured and transported which in turn increase the levels of greenhouse gases from manufacturing factories and vehicles. According to the EPA, the majority of greenhouse gas emissions come from transportation (29% of total emissions) and industry (23%). Therefore, the purpose of this research study is to figure out if solar panels indeed have helped reduce carbon dioxide and greenhouse emissions. We hope to construct a model that shows how many solar panels each city contains and compare it to the data of carbon dioxide emissions for those cities. <br>
For the purpose of this study, although greenhouse gases actually contain carbon dioxide, methane, nitrous oxide, water vapor, and fluorinated gases, we will interchangeably use the terms “greenhouse gases”, “carbon emissions”, “carbon dioxide”, etc.  <br>

#### Prior Work:
According to a study done by Theocharis Tsoutsos, Niki Frantzeskaki, and Vassilia Gekas, from the Department of Environmental Engineering at the Technical University of Crete, solar energy technologies have helped in the reduction of the emission of greenhouse gases with about 1.4 kg/kWh per year. They do mention, however, that there has been a concern on the land use in the case of large-scale systems because in order for large solar panels to be constructed there needs to be a loss of natural habitats<sup>6</sup>. This relates to our study because the creation of these large areas requires a lot of manufacturing and therefore the release of greenhouse gases. <br>
In another study conducted by Giorgio Schileo and Giulia Grancini from the Royal Society of Chemistry, they argue that the advantages of solar panels outweigh the risks associated with manufacturing if you take the necessary precautions with encapsulation and recycling. They suggested that the only toxic material contains in photovoltaic cells (solar panels) is lead, which can be easily substituted<sup>7</sup>. <br>

### Hypothesis <br>
In our study, we would like to explore the correlation between the number of solar panel installations in a city with the carbon dioxide emission. Our hypothesis for the research question is that cities with high numbers of solar panel installations will also have high carbon dioxide emissions. Superficially, this may seem contradictory to the fact that solar panels and other forms of renewable energy reduce the need for the burning of fossil fuels to produce electricity. However, cities with more solar panels are more affluent, meaning that the standard of living is higher so there are more cars per household as well as a higher population, leading to an increase in carbon dioxide emissions through the burning of fossil fuels. <br>
### Data <br>
